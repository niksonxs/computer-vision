```python
from skimage import io
from skimage.metrics import structural_similarity as ssim
```

```python
img=io.imread('lenna noisy.png')
```

```python
io.imshow(img)
```

```python
from matplotlib import pyplot as plt
import cv2
title=['Original', 'Median filter', 'Gaussian filter', 'Bilateral filter']
ax = [plt.subplot(1,4,i+1) for i in range(4)]
images=[img,cv2.medianBlur(img,3),cv2.GaussianBlur(img,(5,5),cv2.BORDER_DEFAULT),cv2.bilateralFilter(img,3,75,75)]
for a in range(4):
    plt.subplot(1,4,a+1)
    ax[a].set_aspect('auto')
    io.imshow(images[a])
    ax[a].axis('off')
    ax[a].set_title(title[a], size=20)
    ax[a].text(0.5,-0.1, ssim(img,images[a],data_range=images[a].max()-images[a].min(),multichannel=True), size=13, ha='center',
         transform=ax[a].transAxes)
plt.subplots_adjust(right=3,wspace=0, hspace=0)
```
