```python
from skimage import io
import numpy as np
import cv2
```

```python
rgb=io.imread("catpat.jpg")
```

```python
io.imshow(rgb)
```

```python
def rgb2hsv(rgb):
    h=0
    s=0
    v=0

    rgb= cv2.cvtColor(rgb,cv2.COLOR_RGBA2RGB) #in caz ca uneori imaginea se citeste ca RGBA
    hsv= np.zeros(rgb.shape)
    for x in range(rgb.shape[0]):
        for y in range(rgb.shape[1]):
            pixel= [color / 255 for color in rgb[x,y]]
            cmax=max(pixel)
            cmin=min(pixel)
            delta=cmax-cmin
            if delta==0:
                h=0
            elif cmax==pixel[0]:
                h=60*(((pixel[0]-pixel[2])/delta)%6)
            elif cmax==pixel[1]:
                h=60*(((-pixel[0]+pixel[2])/delta)+2)
            elif cmax==pixel[2]:
                h=60*(((pixel[0]-pixel[1])/delta)+4)
            v=cmax
            if cmax:
                s=delta/cmax
            else:
                s=0
            hsv[x,y]=[int(h/2),int(s*255),int(v*255)]
    hsv=hsv.astype(int)
    return hsv
```

```python
hsv=rgb2hsv(rgb)
io.imshow(hsv)
```
