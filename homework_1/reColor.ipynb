```python
from skimage import io
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = io.imread('c://cat.png')
```

```python
io.imshow(img)
```

```python
def reColor(img,contrast,brightness):
    factor = (259 * (contrast + 255)) / (255 * (259 - contrast))
    new_img=np.zeros(img.shape)
    new_img=new_img.astype(int)
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            for c in range(img.shape[2]):
                new_img[x,y,c]=np.clip((int(factor * (img[x,y,c] - contrast) + contrast))+brightness,0,255)
    
    return new_img
```

```python
def contrast(img,contrast):
    factor = (259 * (contrast + 255)) / (255 * (259 - contrast))
    new_img=np.zeros(img.shape)
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            for c in range(img.shape[2]):
                new_img[x,y,c]=int(factor * (img[x,y,c] - contrast) + contrast)
    new_img=new_img.astype(int)
    return new_img
```

```python
io.imshow(reColor(img,100,240))
```

```python
ax = [plt.subplot(5,5,i+1) for i in range(25)]
data=[10,-10,10,-5,10,0,10,5,10,10,  5,-10,5,-5,5,0,5,5,5,10,  0,-10,0,-5,0,0,0,5,0,10, -5,-10,-5,-5,-5,0,-5,5,-5,10, -10,-10,-10,-5,-10,0,-10,5,-10,10]
for a in range(25):
    plt.subplot(5,5,1+a)
    ax[a].axis('off')
    ax[a].set_aspect('auto')
    io.imshow(reColor(img,data[2*a],data[2*a-1]))
plt.subplots_adjust(wspace=0, hspace=0)
```
